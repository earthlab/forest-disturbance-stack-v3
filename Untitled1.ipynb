```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# File path
fname = "data/derived/annual_stacks/binary/annual_stack_any_2000.tif"

# Open raster
with rasterio.open(fname) as src:
    print(f"Raster shape: ({src.count}, {src.height}, {src.width})")
    print(f"Number of bands: {src.count}")
    
    band_names = ["wf","bt","hd","pd"]
    
    for i, name in enumerate(band_names, start=1):
        band = src.read(i)
        unique_vals = np.unique(band)
        n_total = band.size
        n_ones = np.sum(band == 1)
        n_zeros = np.sum(band == 0)
        n_nodata = np.sum(band == src.nodata)
        
        print(f"\nBand {i} ({name}):")
        print(f"  Unique values: {unique_vals}")
        print(f"  Min={band.min()}, Max={band.max()}")
        print(f"  1s={n_ones}, 0s={n_zeros}, NoData={n_nodata}, Total pixels={n_total}")
        
        # Quick visual check
        plt.imshow(band, cmap='viridis')
        plt.title(f"{name} - 2000")
        plt.colorbar(label='0=No, 1=Yes')
        plt.show()
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# File path
fname = "data/derived/annual_stacks/binary/annual_stack_any_2020.tif"

band_names = ["wf", "bt", "hd", "pd"]

with rasterio.open(fname) as src:
    nodata = src.nodata

    print(f"Raster shape: ({src.count}, {src.height}, {src.width})")
    print(f"Number of bands: {src.count}")
    print(f"NoData value: {nodata}")

    for i, name in enumerate(band_names, start=1):
        band = src.read(i)

        # --- Stats (explicit, robust)
        unique_vals = np.unique(band)
        n_total = band.size
        n_nodata = np.sum(band == nodata)
        n_valid = n_total - n_nodata
        n_ones = np.sum(band == 1)
        n_zeros = np.sum(band == 0)

        print(f"\nBand {i} ({name}):")
        print(f"  Unique values: {unique_vals}")
        print(f"  Min={band.min()}, Max={band.max()}")
        print(f"  1s={n_ones:,}")
        print(f"  0s={n_zeros:,}")
        print(f"  NoData={n_nodata:,}")
        print(f"  Valid pixels={n_valid:,}")
        print(f"  % disturbed (of valid) = {100 * n_ones / n_valid:.2f}%")

        # --- Mask NoData for plotting
        band_masked = np.ma.masked_equal(band, nodata)

        # --- Plot
        # Mask nodata
        masked = np.ma.masked_equal(band, nodata)

        # Subsample for plotting if needed
        masked_plot = masked[::20, ::20]

        # Define a custom colormap: 0=dark, 1=red
        cmap = mcolors.ListedColormap(['black', 'red'])

        plt.figure(figsize=(8, 10))
        plt.imshow(masked_plot, cmap=cmap, interpolation='nearest')
        plt.title(f"{name.upper()} binary (2000)")
        plt.colorbar(ticks=[0,1], label='0=No, 1=Yes')
        plt.show()

```
