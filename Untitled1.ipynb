```python
import rasterio
from pathlib import Path

root = Path("data/derived/annual_stacks")
years = range(2000, 2021)

for yr in years:
    fname = root / f"annual_stack_{yr}.tif"
    with rasterio.open(fname) as src:
        print(f"{fname}: {src.count} bands")
```

```python
import numpy as np

with rasterio.open(fname) as src:
    for i, name in enumerate(["wf", "bt", "hd", "pd"], start=1):
        band = src.read(i)
        print(f"Year {yr}, band {i} ({name}): min={band.min()}, max={band.max()}")
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# File path
fname = "data/derived/annual_stacks/binary/annual_stack_any_2000.tif"

# Open raster
with rasterio.open(fname) as src:
    print(f"Raster shape: ({src.count}, {src.height}, {src.width})")
    print(f"Number of bands: {src.count}")
    
    band_names = ["wf","bt","hd","pd"]
    
    for i, name in enumerate(band_names, start=1):
        band = src.read(i)
        unique_vals = np.unique(band)
        n_total = band.size
        n_ones = np.sum(band == 1)
        n_zeros = np.sum(band == 0)
        n_nodata = np.sum(band == src.nodata)
        
        print(f"\nBand {i} ({name}):")
        print(f"  Unique values: {unique_vals}")
        print(f"  Min={band.min()}, Max={band.max()}")
        print(f"  1s={n_ones}, 0s={n_zeros}, NoData={n_nodata}, Total pixels={n_total}")
        
        # Quick visual check
        plt.imshow(band, cmap='viridis')
        plt.title(f"{name} - 2000")
        plt.colorbar(label='0=No, 1=Yes')
        plt.show()
```
